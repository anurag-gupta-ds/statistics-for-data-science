```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
dataset=[10,10,11,12,13,15,16,16,18,10,9,9,12,12,13,14,14,18,18,19,20,18,17,105,10,110]
plt.hist(dataset)
```

```python
outliers = []

def detect_outliers(data):
    threshold = 3
    mean = np.mean(data)
    std = np.std(data)
    
    for i in data:
        z_score = (i - mean) / std
        if np.abs(z_score) > threshold:
            outliers.append(i)
    
    return outliers

detect_outliers(dataset)
```

```python
import numpy as np

def detect_outliers_iqr(data):
    outliers = []
    q1, q3 = np.percentile(data, [25, 75])
    iqr = q3 - q1
    lower_fence = q1 - 1.5 * iqr
    upper_fence = q3 + 1.5 * iqr

    for i in data:
        if i < lower_fence or i > upper_fence:
            outliers.append(i)

    return outliers
detect_outliers_iqr(dataset)
```
